<center><h3>Service Fix Cost Savings Summary</h3></center>

## Background Information:

HNA Market Quality Denso team members identified market problem where CRV heater cores were clogged, causing no heat from driver's side vents.  To fix, dealers have been replacing the heater core.  Replacing the heater core on average costs <strong>$592</strong> per claim.

## Current warranty trend and cost when replacing the heater core:

<img src="United_Wayne_Campaign/cum_defect.png" />
<img src="United_Wayne_Campaign/cost.png" />

#### This market failure has a high durability trend and warranty costs that exceed $3 million dollars.

## Service Fix Procedure:

Instead of replacing the heater core assembly which requires extensive labor, HNA Denso team members proposed intead that dealers just flush the cooling system with C-L-R mixture.  The average warranty cost associated with just flushing the cooling system is &#36;184.  Relative to replacing the heater core, this equates to a cost savings of approximately <strong>$408</strong> per claim.

## Service Fix Cost Savings:

```python
import pyodbc, crypto, codecs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.dates import YearLocator, MonthLocator, WeekdayLocator, DayLocator, DateFormatter
from matplotlib.ticker import MultipleLocator
import dateutil
from matplotlib.ticker import MultipleLocator

# Get password and unencrypt it
key = codecs.escape_decode(open(r'D:\webapps\_server\pyodbc\key', 'r').read())[0]
pw_file = open(r'D:\webapps\_server\pyodbc\cmq.txt', 'r')
pwd = pw_file.read()
pw = crypto.decrypt_with_key(key, pwd)
userid = 'rb10'
pw_file.close()

cnxn_string = 'DSN=CMQ_PROD;UID=' + userid + ';PWD=' + pw

cnxn = pyodbc.connect(cnxn_string)
cursor = cnxn.cursor()

#sql = open(r'\\mapfile01\aqg\mq\group\Information\ipython\coremq\heater_core.sql','r').read()
sql = """
/** Claims set **/
WITH CLAIMS AS (
SELECT
CLAIMS.VIN_SK,
CLAIMS.CLM_SK

FROM
CMQ.V_FACT_CLM_TRX CLAIMS

INNER JOIN CMQ.V_DIM_MTO MTO
ON CLAIMS.MTO_SK = MTO.MTO_SK

INNER JOIN CMQ.V_DIM_PART PART
ON CLAIMS.FAIL_PART_SK = PART_SK

WHERE
MTO.MDL_YR IN( 2012,2013,2014 )
AND MTO.FCTRY_CD IN('ELP','HDM','HCM')
AND MTO.MDL_NM = 'CRV'
AND MTO.DEST_CD IN('KA','KL')
AND PART.SHRT_PART_NO = '79115'

),

/** CLR only claims **/
REPLACED_CLR AS (
SELECT DISTINCT
CLM_SK,
REPLPART.PART_NO,
REPLPART.UNFRMT_PART_NO,
REPLPART.SHRT_PART_NO,
REPL.REPL_PART_QTY,
REPL.ADJ_PART_QTY

FROM
CMQ.V_FACT_CLM_REPL_PART REPL
INNER JOIN CMQ.V_DIM_PART REPLPART
ON REPL.REPL_PART_SK = REPLPART.PART_SK

WHERE
(REPLPART.UNFRMT_PART_NO LIKE '0L999%'
 OR REPLPART.UNFRMT_PART_NO LIKE 'OL999%'
)
AND REPL.ADJ_PART_QTY > 0
AND REPL.REPL_PART_QTY > 0

),

/** Heater core replacement claims **/
REPLACED_CORE AS (
SELECT DISTINCT
CLM_SK,
REPLPART.PART_NO,
REPLPART.UNFRMT_PART_NO,
REPLPART.SHRT_PART_NO,
REPL.REPL_PART_QTY,
REPL.ADJ_PART_QTY

FROM
CMQ.V_FACT_CLM_REPL_PART REPL

INNER JOIN CMQ.V_DIM_PART REPLPART
ON REPL.REPL_PART_SK = REPLPART.PART_SK

WHERE
REPLPART.UNFRMT_PART_NO LIKE '79115%'
AND REPL.ADJ_PART_QTY > 0
AND REPL.REPL_PART_QTY > 0

)

/** Main SELECT statement **/
SELECT
CLAIMS.CLM_SK,
MTO.MDL_YR AS YEAR,
MTO.FCTRY_CD AS FACTORY,
MTO.MDL_NM AS MODEL,
RECDT.CAL_DT AS REC_DATE,
RECDT.CAL_YR_MO_FMT_NO AS REC_YR_MTH

FROM CLAIMS INNER JOIN REPLACED_CLR ON
CLAIMS.CLM_SK = REPLACED_CLR.CLM_SK

LEFT JOIN REPLACED_CORE on
CLAIMS.CLM_SK = REPLACED_CORE.CLM_SK

INNER JOIN CMQ.V_FACT_CLM_TRX FCLAIM ON
CLAIMS.CLM_SK = FCLAIM.CLM_SK

INNER JOIN CMQ.V_DIM_DATE RECDT ON
FCLAIM.CMQ_RCV_DT_SK = RECDT.CAL_DT_SK

INNER JOIN CMQ.V_DIM_MTO MTO ON
FCLAIM.MTO_SK = MTO.MTO_SK

WHERE
REPLACED_CORE.CLM_SK IS NULL

ORDER BY RECDT.CAL_DT
"""

df = pd.read_sql(sql, cnxn, index_col=None, parse_dates={'REC_DATE':"%Y-%m-%d"})

# Close connections
cursor.close()
cnxn.close()
```

```python
clr = df.pivot_table(values='CLM_SK',index='REC_DATE',aggfunc='count')
clr_df = pd.DataFrame(clr.values, index=clr.index, columns=['QTY'])
clr_df['COST'] = clr_df.QTY * 408
clr_df['CUM_CST'] = clr_df['COST'].cumsum()

bymonth = df.pivot_table(values='CLM_SK',index='REC_YR_MTH',aggfunc='count')
bymonth_df = pd.DataFrame(bymonth.values, index=bymonth.index, columns=['QTY'])

rng = pd.date_range(start='01/01/2013', end='11/30/2014', freq='M')
xlabels = [value.strftime('%Y-%m') for value in rng]
bymonth_df = bymonth_df.reindex(xlabels, fill_value=0)
bymonth_df['COST'] = bymonth_df.QTY * 408
bymonth_df['CUM_CST'] = bymonth_df['COST'].cumsum()
```

```python
fig, axes = plt.subplots(2,1)

month  = MonthLocator(bymonth=range(1,13),  bymonthday=1, interval=1)
week   = WeekdayLocator(byweekday=dateutil.rrule.MO)  # Every MOnday or WeekdayLocator(byweekday=0) works too
monthsFmt = DateFormatter("%b '%y")

axes[0].xaxis.set_major_locator(month)
axes[0] = fig.add_axes([0, 1.4, 1, 1])  # left, bottom, width, height
axes[0].plot_date(clr_df.index, clr_df['CUM_CST'].values,'r')
y_major_ticks = MultipleLocator(5000)
axes[0].yaxis.set_major_locator(y_major_ticks)
axes[0].set_title("Cumulative Cost Savings To Date: " + str(clr_df['CUM_CST'][-1:].values[0]))
axes[0].yaxis.set_ticks_position('right')
axes[0].yaxis.set_label_position('right')
axes[0].set_ylabel("US$")
plt.setp(axes[0].get_xticklabels(), rotation=-90)

axes[1] = fig.add_axes([0, 0, 1, 1])
y_major_ticks = MultipleLocator(10)
axes[1].yaxis.set_major_locator(y_major_ticks)
plt.bar(np.arange(len(bymonth_df.index)), bymonth_df['QTY'].values, align='center')
axes[1].xaxis.set_major_locator(month)
axes[1].yaxis.set_label_position('right')
axes[1].yaxis.set_ticks_position('right')
axes[1].set_ylabel("QTY")
axes[1].xaxis.set_major_formatter(monthsFmt)
axes[1].xaxis.set_major_locator(month)
axes[1].set_title("QTY of Claims")
axes[1].set_xticks(np.arange(len(bymonth_df.index)))
axes[1].set_xticklabels(bymonth_df.index)
plt.setp(axes[1].get_xticklabels(), rotation=-90)
plt.show()
```
