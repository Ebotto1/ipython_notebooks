```python
import pandas.io.data as web
from datetime import datetime
import matplotlib.pyplot as plt

ticker = 'HMC'
start = datetime(2014, 10, 1)
end = datetime(2014, 10, 31)

f=web.DataReader(ticker, 'yahoo', start, end)

f["Adj Close"].plot()
xticks(rotation=270)
title("Adj Close for Ticker: " + ticker)
show()
```

```python
from ggplot import *
import pandas.io.data as web
from datetime import datetime
import pandas as pd

ticker = 'HMC'
start = datetime(2014, 10, 1)
end = datetime(2014, 10, 31)

df=web.DataReader(ticker, 'yahoo', start, end)
```

```python
df.reset_index(level=0, inplace=True)
```

```python
df.head()
```

```python
df.drop('Volume', axis=1, inplace=True)
```

```python
df.head()
```

```python
ggplot(df, aes(x='Date')) + \
    geom_line(aes(y = 'Open', color='red')) + \
    geom_line(aes(y = 'Close', color='blue')) + \
    xlab('DATE') + ylab('Open or Close')
```

```python
df_long = pd.melt(df, id_vars=['Date'])
```

```python
df_long
```

```python
ggplot(df_long, aes(x='Date', y='value', colour='variable')) + geom_line() + ggtitle('Honda Stock')
```
