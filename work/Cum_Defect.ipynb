```python
import pandas as pd
import numpy as np

# Fetch the cumulative defect chart data
cumdef = pd.read_excel(r'\\hdcnas02\AQ_MarketQuality\DensoOBD\WarrantySummaries\79160 - Air Mix Motor\11M_Civic_79160.xlsx','ChartData')

# Chart data is initially grouped across rows, but should be grouped in columns, so we need to transpose
values = np.transpose(cumdef.values) * 100.00

# Get column headers
headers = cumdef.columns
```

```python
# Create list of plot markers
from itertools import cycle
lines = ["-","--","-.",":"]
linecycler = cycle(lines)

# Create a dictionary from the values and headers lists
for data, txt in(zip(values, headers)):
    plot(data,next(linecycler),label=txt)
    
legend(bbox_to_anchor=(1.5,1.1))
title("09-13M Civic 79160 Replaced")
ylabel("Cumulative Def %")
xlabel("MFSS")
grid()
show()
```
