```python
import pyodbc, base64
import pandas
import numpy as np

# Get password and "de-fuzzy" it
pw_file = open(r'D:\webapps\_server\pyodbc\cmq.txt', 'r')
pw = base64.b64decode(pw_file.read())
userid = 'rb10'
pw_file.close()

cnxn_string = 'DSN=CMQ_PROD;UID=' + userid + ';PWD=' + pw

cnxn = pyodbc.connect(cnxn_string)
cursor = cnxn.cursor()

sql = """
WITH claims_temp AS (
SELECT
CLAIMS.MTO_SK,
CLAIMS.CLM_SK,
CLAIMS.VIN_SK,
FEATURES.MDL_YR AS MODEL_YEAR,
FEATURES.FCTRY_CD AS FACTORY_CODE,
FEATURES.MDL_NM AS MODEL_NAME,
PART.SHRT_PART_NO||' - '||SHRT_PART.PART_DESC AS PART5_NAME,
PART.SHRT_PART_NO AS PART5,
SHRT_PART.PART_DESC AS PART_DESC,
PART_GRP.PART_GRP_NM,
SUBSTR(LABOR.LAB_OPRTN_CD,1,1) AS BLO1,
CBU.CBU_LOC_NM,
FEATURES."ENGINE CYLINDERS" AS ENG_CYL,
FEATURES."DOORS" AS DOOR_QTY,
LABOR.LAB_OPRTN_CD AS PRILABCD,
CMQ_CAL.CAL_DT as CMQ_RECV_DATE,
RO_CAL.CAL_DT AS RODATE,
RO_CAL.CAL_YR_MO_FMT_NO AS RO_YR_MTH,
CLM_DETAILS.ORIG_CUST_CNTN_TXT,
CLM_DETAILS.DLR_NO AS CLAIM_DEALERNO,
CLM_DETAILS.CLM_NO AS CLAIM_NO,
CLM_DETAILS.RO_YR AS CLAIM_RO_YR,
CLAIMS.USD_REMB_CLM_AMT AS TOTALCHG,
SYMP_CLASS.SYMP_CLASS_NM,
PD.PROB_DEFN_NM

FROM CMQ.V_FACT_CLM_TRX CLAIMS

INNER JOIN CMQ.V_DIM_CLM CLM_DETAILS
ON CLAIMS.CLM_SK = CLM_DETAILS.CLM_SK

INNER JOIN CMQ.V_FACT_CLM_ENRICH CLM_ENRICH
ON CLAIMS.CLM_SK = CLM_ENRICH.CLM_SK

LEFT JOIN CMQ.V_FACT_PD_CLM_ENRICH PD_ENRICH
ON CLAIMS.CLM_SK = PD_ENRICH.CLM_SK

LEFT JOIN CMQ.V_DIM_PROB_DEFN PD
ON PD_ENRICH.PROB_DEFN_SK = PD.PROB_DEFN_SK

LEFT JOIN CMQ.V_DIM_PART_GRP PART_GRP
ON CLM_ENRICH.PART_GRP_SK = PART_GRP.PART_GRP_SK

LEFT JOIN CMQ.V_DIM_SYMP_CLASS SYMP_CLASS
ON CLM_ENRICH.SYMP_CLASS_SK = SYMP_CLASS.SYMP_CLASS_SK

INNER JOIN CMQ.V_DIM_PART PART
ON CLAIMS.FAIL_PART_SK = PART.PART_SK

INNER JOIN CMQ.V_DIM_SHRT_PART SHRT_PART
ON PART.SHRT_PART_NO = SHRT_PART.SHRT_PART_NO

INNER JOIN CMQ.V_DIM_MTO MTO
ON CLAIMS.MTO_SK = MTO.MTO_SK

INNER JOIN CMQ.V_DIM_DLR CLAIM_DLR
ON CLAIMS.SVC_DLR_SK = CLAIM_DLR.DLR_SK

INNER JOIN CMQ.V_DIM_DFCT_CD DFCT_CD
ON CLAIMS.DFCT_CD_SK = DFCT_CD.DFCT_CD_SK

INNER JOIN CMQ.V_DIM_DATE RO_CAL
ON CLAIMS.RO_DT_SK = RO_CAL.CAL_DT_SK

INNER JOIN CMQ.V_DIM_LAB_OPRTN_CD LABOR
ON CLAIMS.PRM_LAB_OPRTN_SK = LABOR.LAB_OPRTN_SK

INNER JOIN CMQ.V_DIM_SUBL_CD SUBLET1
ON CLAIMS.SUBL_01_SK = SUBLET1.SUBL_CD_SK

INNER JOIN CMQ.V_DIM_SUBL_CD SUBLET2
ON CLAIMS.SUBL_02_SK = SUBLET2.SUBL_CD_SK

INNER JOIN CMQ.V_DIM_MTO_FEATURE_PIVOT FEATURES
ON CLAIMS.MTO_SK = FEATURES.MTO_SK

INNER JOIN CMQ.V_DIM_DATE CMQ_CAL
ON CLAIMS.CMQ_RCV_DT_SK = CMQ_CAL.CAL_DT_SK

INNER JOIN CMQ.V_DIM_PART_CBU_LOC_ASSOC CBU
ON CLAIMS.FAIL_PART_SK = CBU.PART_SK


WHERE
MTO.MDL_YR IN(2012, 2013)
AND MTO.FCTRY_CD IN( 'MAP','ELP', 'HMIN', 'HDM' )
AND MTO.MDL_NM IN( 'ACCORD','TL','RDX','CIVIC','CRV','CROSSTOUR', 'ILX' )
AND CLAIM_DLR.STATE_CD IN ('AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA', 'ID', 'IL', 'IN',
'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH',
'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY')

AND CMQ_CAL.CAL_DT >= '1990-01-01'

/** Exclude Campaigns, transport claims, and goodwill **/
AND CLAIMS.CAMP_SK <= 0
AND CLAIMS.REMB_PART_AMT >= 0

AND (LABOR.LAB_OPRTN_CD LIKE '6%'
  OR (LABOR.LAB_OPRTN_CD LIKE '7%'
    AND NOT(
      CBU.CBU_LOC_NM LIKE '%EXTER%'
      OR CBU.CBU_LOC_NM LIKE '%INTERIOR%'
    )
  )
  OR CBU.CBU_LOC_NM LIKE '%OBD%'
  OR PART_GRP.PART_GRP_NM LIKE '%TPMS%'
  OR PART_GRP.PART_GRP_NM LIKE '%WIPER%'
  OR PART_GRP.PART_GRP_NM LIKE '%WASHER%'
  OR PART_GRP.PART_GRP_NM LIKE '%SRS%'
  OR LABOR.LAB_OPRTN_CD LIKE '823505%'
  OR PART.SHRT_PART_NO LIKE '1901%'
  OR PART.SHRT_PART_NO LIKE '3935%'
  OR PART.SHRT_PART_NO LIKE '42753%'
  OR PART.SHRT_PART_NO LIKE '06421%'
  OR PART.SHRT_PART_NO LIKE '766%'
  OR PART.SHRT_PART_NO LIKE '768%'
  OR PART.SHRT_PART_NO LIKE '38515%'
  OR SHRT_PART.PART_DESC LIKE '%MICROPHONE%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%SRS%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%S_R_S%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%WEIGHT SENS%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%BAG%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%TPMS%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%HOMELINK%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%HOME LINK%'
  OR CLM_DETAILS.ORIG_CUST_CNTN_TXT LIKE '%GARAG%'
)
AND NOT(LABOR.LAB_OPRTN_CD LIKE '710%'
   OR LABOR.LAB_OPRTN_CD LIKE '711%'
   OR LABOR.LAB_OPRTN_CD LIKE '747%'
   OR LABOR.LAB_OPRTN_CD LIKE '737%'
   OR LABOR.LAB_OPRTN_CD LIKE '0%'
   OR LABOR.LAB_OPRTN_CD LIKE '125517%'  /* Exclude PCM updates */
   OR PART.SHRT_PART_NO LIKE '3979%'
   OR PART.SHRT_PART_NO LIKE '77900%'
   OR PART.SHRT_PART_NO LIKE '0481%'
   OR SHRT_PART.PART_DESC LIKE '%BULB%'
   OR SHRT_PART.PART_DESC LIKE '%HARN%'
   OR PART_GRP.PART_GRP_NM LIKE '%WINDSHIELD%'
   OR PART_GRP.PART_GRP_NM LIKE '%COMBIMETER%'
   OR PART_GRP.PART_GRP_NM LIKE '%CHARGING SYSTEM%'
)



AND NOT(
    LEFT(CLM_DETAILS.CLM_NO,1) IN('T','R')
    OR PART.SHRT_PART_NO LIKE '08%'
    OR SUBLET1.SUBL_CD = 'GW'
    OR SUBLET2.SUBL_CD = 'GW'
)

),

sales_temp as (
SELECT
PROD.VIN_SK,
AF_CAL.CAL_YR_MO_FMT_NO as AF_YR_MTH

FROM
CMQ.V_FACT_VHCL_PRDN_SLS PROD

INNER JOIN CMQ.V_DIM_MTO MTO
ON PROD.MTO_SK = MTO.MTO_SK

INNER JOIN CMQ.V_DIM_DATE AF_CAL
ON PROD.AF_OFF_DT_SK = AF_CAL.CAL_DT_SK

WHERE
MTO.MDL_YR IN(2012, 2013)
AND MTO.FCTRY_CD IN( 'MAP','ELP', 'HMIN', 'HDM' )
AND MTO.MDL_NM IN( 'ACCORD','TL','RDX','CIVIC','CRV','CROSSTOUR', 'ILX' )

)

SELECT
MODEL_YEAR,
FACTORY_CODE,
MODEL_NAME,
PART5_NAME,
PART_GRP_NM,
AF_YR_MTH,
RO_YR_MTH,
BLO1,
ENG_CYL,
DOOR_QTY,
CBU_LOC_NM,
SYMP_CLASS_NM,
PROB_DEFN_NM,
COUNT(*) AS QTY

FROM (

SELECT
claims.*,
sales.*

FROM claims_temp claims

INNER JOIN sales_temp sales
ON claims.VIN_SK = sales.VIN_SK

) AS TEMP

GROUP BY
MODEL_YEAR,
FACTORY_CODE,
MODEL_NAME,
PART5_NAME,
PART_GRP_NM,
AF_YR_MTH,
RO_YR_MTH,
BLO1,
ENG_CYL,
DOOR_QTY,
CBU_LOC_NM,
SYMP_CLASS_NM,
PROB_DEFN_NM"""

try:
    cursor.execute(sql)
except:
    print "There was an error in your SQL"
    
resultset = cursor.fetchall()

# Get column names of the resultset
column_name_list = [tuple[0] for tuple in cursor.description]

# Close connections
cursor.close()
cnxn.close()

# Create a dataframe
df = pandas.DataFrame(resultset,columns=column_name_list)

# Print the dataframe to see what's inside
print df.to_string()
```
