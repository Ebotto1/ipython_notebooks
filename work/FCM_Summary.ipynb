<h2 align="center">Script that summarizes Front Camera Angles</h2>

```python
import sqlite3                         # Needed to access Sqlite3 database
import pandas.io.sql as sqlio          # Makes creation of dataframes/datasets from sql connection very easy
import numpy as np                     # Contains useful mathematical or matrix-related functions

con = sqlite3.connect(r'd:\webapps\denso\server\db\denso')

# A VIN can have more than one LET reads.  We want to just get the lastest LET read.
sql = """
SELECT
txtVIN,
pitch,
roll,
yaw,
date(max(datTime)) as testDate,
txtMY,
txtModel,
txtPlant,
strftime('%Y-%m',date(max(datTime))) as testYrMonth

FROM fcm_camera

WHERE
txtMY = 'D'
and txtPlant = 'A'
and txtModel in('CR','CT')

group by
txtVIN"""

df = sqlio.read_frame(sql, con)

con.close()

# Let's see what the dataframe looks like.  This prints out the first 5 records in our dataframe:
print df.head()
```

#### Printing the mean for each camera angle by LET test month:

```python
print "---------------------------------------------------"
print "Mean Pitch Angle(deg):"
print df.groupby(['testYrMonth'])['pitch'].mean()
print "---------------------------------------------------"
print "Mean Yaw Angle(deg):"
print df.groupby(['testYrMonth'])['yaw'].mean()
print "---------------------------------------------------"
print "Mean Roll Angle(deg):"
print df.groupby(['testYrMonth'])['roll'].mean()
```

#### Create Box Plot of Front Camera Angles and print summary statistics

```python
pitch_values = df['pitch'].values
yaw_values   = df['yaw'].values
roll_values  = df['roll'].values

fig1 = figure(1, facecolor="white")
subplot(1,1,1, axisbg='#cdc9c9')
ax = gca()

boxplot([pitch_values, yaw_values, roll_values], vert=0, whis=3)
title("Box Plot of 2GA Front Camera Angles", weight='bold')
plt.xlabel("Camera Angle in Degrees")
ax.set_yticklabels(['Pitch Angle', 'Yaw Angle', 'Roll Angle'])
ax.xaxis.grid(which='major', linestyle='--', linewidth=1)
major_ticks = MultipleLocator(1)
ax.xaxis.set_major_locator(major_ticks)
show()

print "--------  Roll Angle(deg)  --------"
print "Max:", roll_values.max()
print "Min:", roll_values.min()
print "Mean:", roll_values.mean()
print "Median:", np.median(roll_values)

print "--------  Yaw Angle(deg)  --------"
print "Max:", yaw_values.max()
print "Min:", yaw_values.min()
print "Mean:", yaw_values.mean()
print "Median:", np.median(yaw_values)

print "--------  Pitch Angle(deg)  --------"
print "Max:", pitch_values.max()
print "Min:", pitch_values.min()
print "Mean:", pitch_values.mean()
print "Median:", np.median(pitch_values)
```
