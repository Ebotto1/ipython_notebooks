```python
import pandas as pd
from ggplot import *

df = pd.read_csv('/home/pybokeh/Dropbox/python/ipython/notebooks/pandas/claims_sales.csv', index_col=None)
```

```python
df = df[df['RO_MTH']!='9999-12']
df
```

```python
df['LABEL'] = df['MDL_YR'].map(str) + '-' + df['MDL_NM'].map(str) + '-' +df['FCTRY_CD'].map(str)
df
```

```python
dfTemp = df.copy()
```

```python
cum_sales = pd.pivot_table(dfTemp,values='SALE_QTY',index='RO_MTH',columns=['LABEL'],aggfunc=np.sum).cumsum()
```

```python
categories = cum_sales.columns
```

```python
cum_sales
```

```python
cum_sales.columns = [column + ' Cum Sales' for column in cum_sales.columns]
cum_sales
```

```python
cum_claims = pd.pivot_table(dfTemp,values='CLAIM_QTY',index='RO_MTH',columns=['LABEL'],aggfunc=np.sum).cumsum()
cum_claims
```

```python
cum_claims.columns = [column + ' Cum Claims' for column in cum_claims.columns]
cum_claims
```

```python
left_joined = pd.merge(cum_sales,cum_claims,how='left',left_index=True,right_index=True)
left_joined
```

## Dynamically-generate new column names and their corresponding defect rates

```python
i = 0
while i < len(categories):
    for label in categories:
        left_joined[label + ' Cum Defect Rate'] = left_joined.iloc[:,i+2] / left_joined.iloc[:,i] * 100
        i = i + 1
left_joined
```

```python
defect_rates = left_joined.filter(like='Defect')  # create a dataframe with columns containing the word 'Defect'
defect_rates
```

```python
defect_rates.reset_index(level=0, inplace=True)  # Convert index to column
defect_rates
```

```python
df_long = pd.melt(defect_rates, id_vars=['RO_MTH']).dropna()
df_long.head()
```

```python
df_long.rename(columns={'variable':'YR-MDL-FAC','value':'Def_%'}, inplace=True)
df_long.head()
```

```python
grouped = df_long['YR-MDL-FAC'].value_counts()
new_index_list = [[value for value in range(index)] for index in grouped.values]

new_index = []
for mylist in new_index_list:
    new_index = new_index + mylist
    
df_long.index = new_index
```

```python
df_long
```

```python
df_long.reset_index(level=0, inplace=True)
df_long.rename(columns={'index':'MFSS'}, inplace=True)
df_long
```

```python
from ggplot import *

ggplot(df_long, aes(x='MFSS', y='Def_%', colour='YR-MDL-FAC')) + \
        geom_line() + \
        ylab('Cumulative Defect Rate (%)') + \
        ggtitle('Warranty Cumulative Defect Rate')
```

```python
pivoted = df_long.pivot(index='MFSS',columns='YR-MDL-FAC',values='Def_%')
pivoted
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# Create list of plot markers
from itertools import cycle
lines = ["-","--","-.",":"]
linecycler = cycle(lines)

x = pivoted.index
headers = pivoted.columns

i = 0
while i < len(headers):
    plt.plot(x,pivoted.iloc[:,i].values,next(linecycler),label=headers[i])
    i = i + 1

#plt.plot(x,y,label=headers)
plt.legend(bbox_to_anchor=(1.8,1.0))
plt.title('Warranty Cumulative Defect %')
plt.xlabel('MFSS')
plt.ylabel('Cumulative Defect Rate')
plt.grid(True)
plt.show()
```
