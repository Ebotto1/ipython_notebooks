```python
from urllib import request
import pandas
from matplotlib.dates import date2num
from datetime import datetime
import time

yen = request.urlopen('http://research.stlouisfed.org/fred2/series/EXJPUS/downloaddata/EXJPUS.csv')
df = pandas.read_csv(yen, index_col=0)

#yen_dates = [date2num(datetime.strptime(value, '%Y-%m-%d')) for value in df.index]
#yen_tuple = sorted((time.mktime(time.strptime(date, '%Y-%m-%d')), value) for date, value in df.itertuples())
yen_tuple = sorted((date2num(datetime.strptime(date, '%Y-%m-%d')), value) for date, value in df.itertuples())

#yen_generator = ((date,price[0]) for date, price in zip(yen_dates,df.values.tolist()))
```
