```python
import pandas as pd
import seaborn as sb
from IPython.html.widgets import interact
sb.set(font="serif")

df = pd.read_excel(r'D:\temp\battery.xlsx', 'Claims', index_col=None, na_values=['NA'])
df = df[['MODEL_NAME','ENGINE_CYLINDERS','DAYS_TO_FAIL_MINZERO','DAYS_TO_SALE']]

grouped_models = df.groupby('MODEL_NAME').size().order(ascending=False)
models = grouped_models.index.tolist()

grouped_cylinders = df.groupby('ENGINE_CYLINDERS').size().order(ascending=False)
cylinders = grouped_cylinders.index.astype('str').tolist()  # drop down items must be of type str

@interact
def dtf_vs_dts(model=models, cyl=cylinders):
    dataset = df[(df['MODEL_NAME']==model) & (df['ENGINE_CYLINDERS']==int(cyl))]
    try:
        sb.jointplot("DAYS_TO_FAIL_MINZERO","DAYS_TO_SALE",data=dataset, kind='kde', xlim=(-100,800), ylim=(-100,600))
    except:
        print("Not a valid selections")
```
