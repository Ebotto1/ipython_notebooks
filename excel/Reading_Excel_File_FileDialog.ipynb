<h2 align='center'>Python script that opens an Excel file located on a network share</h2>

<h3>Data Retrieval Phase</h3>

<p>Here, we are retrieving the data that is stored in an Excel file using openpyxl module.  With this version, an open file dialog box pops up so that you can navigate to the data you want to retrieve.</p>

```python
import pandas
import openpyxl # need this module if opening Excel 2007+
import statsmodels.api as sm
import itertools
import Tkinter, tkFileDialog


class GetDataSet(Tkinter.Frame):
    def __init__(self, master):
        Tkinter.Frame.__init__(self, root)
        #self.filename = ''
        
        # define options for opening or saving a file
        self.file_opt = options = {}
        options['defaultextension'] = '.xlsx'
        options['filetypes'] = [('all files', '.*'), ('text files', '.txt'), ('< Excel 2007', '.xls'), ('>= Excel 2007', '.xlsx')]
        options['initialdir'] = 'D:\\'
        #options['initialfile'] = 'myfile.txt'
        options['parent'] = root
        options['title'] = 'Get your data set now!'

        # This is only available on the Macintosh, and only when Navigation Services are installed.
        #options['message'] = 'message'

        # if you use the multiple file version of the module functions this option is set automatically.
        #options['multiple'] = 1

        # defining options for opening a directory
        self.dir_opt = options = {}
        options['initialdir'] = 'C:\\'
        options['mustexist'] = False
        options['parent'] = root
        options['title'] = 'Get your data'
        
    def askopenfilename(self):
        """Returns an opened file in read mode.
        This time the dialog just returns a filename and the file is opened by your own code.
        """

        # get filename
        filename = tkFileDialog.askopenfilename(**self.file_opt)
        return filename
        
#if __name__ == "__main__":
root = Tkinter.Tk()
root.title("Get Data Set")
#GetDataSet(root).pack()
app = GetDataSet(root)
data_file = app.askopenfilename()
root.destroy()
root.mainloop()

# Open the Excel file
try:
    xls = pandas.ExcelFile(data_file)
    df = xls.parse("Claims")  # import the Claims sheet, all columns, into a pandas DataFrame
except:
    print "No file selected or there was an error while trying to get your data"

# re-write the dataframe so that it contains only the VIN, model year, factory, model name, MTF, and DTF columns
df = df[["VIN", "MODEL_YEAR", "FACTORY_CODE", "MODEL_NAME","MILES_TO_FAIL","DAYS_TO_FAIL_MINZERO"]]
```
