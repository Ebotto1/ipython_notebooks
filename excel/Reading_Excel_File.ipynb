# Python script that opens an Excel file located on a network share

                                                                      Data Input Phase

```python
import pandas
import openpyxl # need this module if opening Excel 2007+
import matplotlib.pyplot as plt

# Open the Excel file
xls = pandas.ExcelFile(r"\\mapfile01\aqg\mq\group\DOBD\Scorecard data\Jul - 2012 data\TL_72148.xls")
df = xls.parse("Claims")  # import the Claims sheet, all columns, into a pandas DataFrame

# re-write the dataframe so that it contains only the VIN, model year, factory, model name, MTF, and DTF columns
df = df[["VIN", "MODEL_YEAR", "FACTORY_CODE", "MODEL_NAME","MILES_TO_FAIL","DAYS_TO_FAIL_MINZERO"]]

# Let's filter our dataframe to contain just 2009 TL data and get just the MTF values
df_2008 = df[ (df["MODEL_YEAR"]==2008) ]["MILES_TO_FAIL"]
df_2009 = df[ (df["MODEL_YEAR"]==2009) ]["MILES_TO_FAIL"]
df_2010 = df[ (df["MODEL_YEAR"]==2010) ]["MILES_TO_FAIL"]
```

                                                                      Charting Phase

```python
# Create figure and inititialize it
fig1 = plt.figure(1)
fig1.subplots_adjust(hspace=0.3)
fig1.set_size_inches(9,11)

# Plot the histogram
subplot(2,1,1)
hist( (df_2008.values, df_2009.values, df_2010.values), label=("2008","2009","2010"), normed=True, bins=15 )
title("2009 TL Keyless Remote Battery MTF")
xlabel("Miles To Failure")
ylabel("Frequency")
grid(axis="y")
legend()

# Plot the box plot
plt.subplot(2,1,2)
axis = plt.gca()
boxplot((df_2008.values, df_2009.values, df_2010.values), vert=0)
axis.set_yticklabels( ["2008","2009","2010"] )
title("Box Plots")
xlabel("Miles to Failure")
grid(axis="x")

show()

print "Summary statistics for 2008 TL:"
print df_2008.describe()

print "\n"

print "Summary statistics for 2009 TL:"
print df_2009.describe()

print "\n"

print "Summary statistics for 2010 TL:"
print df_2010.describe()
```
