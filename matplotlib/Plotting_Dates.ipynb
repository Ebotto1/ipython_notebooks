```python
import pandas as pd
dates = pd.date_range('1/1/2011', periods=365, freq='D')
```

```python
import numpy as np
```

```python
data = np.random.rand(365)*1000
```

```python
# http://matplotlib.org/api/dates_api.html#module-matplotlib.dates
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, WeekdayLocator, DayLocator, DateFormatter
import dateutil
from matplotlib.ticker import MultipleLocator

fig = plt.figure()
ax = plt.subplot(1,1,1)

# Let's define x-axis date interval sizes: month, week, or day
month_interval = 1   
months = MonthLocator(bymonth=range(1,13),  bymonthday=1, interval=month_interval)
month  = MonthLocator(bymonth=range(1,13),  bymonthday=1, interval=1)
week   = WeekdayLocator(byweekday=dateutil.rrule.MO)  # Every MOnday
day    = DayLocator(bymonthday=range(1,32), interval=1)

# Let's define x-axis formatting intervals
monthsFmt = DateFormatter("%b '%y")
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(monthsFmt)
ax.xaxis.set_minor_locator(week)
ax.xaxis.grid(which='major')

# Let's define y-axis formatting
y_major_ticks = MultipleLocator(50)
ax.yaxis.set_major_locator(y_major_ticks)
ax.yaxis.grid(which='major')

#plt.title("\large{Plotting Dates}")
plt.title("Plotting Dates Example")
plt.plot(dates, data)
plt.xticks(rotation=-90)
plt.xlabel("Major ticks=months\nMinor ticks=weeks")
plt.ylabel("values")
plt.show()
```
