### Histogram

[[Back to Top](#top)]

Example on how to plot data points or values that follow the standard Normal distribution.  
The probability distribution function (pdf) of the standard Normal distribution is:<br>
<center><h2>$f(x,\mu,\sigma)=\frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$</h2></center><br>
where $\mu$ is the *mean* and $\sigma$ is the *standard deviation* of the distribution.

```python
from numpy import random as nprandom
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

mu, sigma = 0, 1
variates = nprandom.normal(mu,sigma,1000)

count, bins, ignored = plt.hist(variates, 30, normed=True) # plots the histogram/bar chart

# plots the pdf of the standard Normal distribution per equation above
axes.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
axes.set_title("Histogram of a Standard Normal Variate", weight='bold')
axes.set_xlabel("bins")
axes.set_ylabel("Frequency")
plt.show()
```

### Box Plot

[[Back to Top](#top)]

```python
from numpy import random as nprandom
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

mu, sigma = 0, 1
variates = nprandom.normal(mu,sigma,1000)

count, bins, ignored = plt.hist(variates, 30, normed=True)

#sns.boxplot(variates, vert=False)  # run this line for a better looking box plot
axes.boxplot(variates, vert=False)
axes.set_xlabel('x')
axis = plt.gca()
axis.yaxis.set_ticklabels(['Normal Variate'])
axes.set_title("Box Plot of a Standard Normal Variate", weight='bold')
plt.show()
```

### Seaborn's Violin Plot (a combination of boxplot and kernel density plot)

[[Back to Top](#top)]

```python
from numpy import random as nprandom
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

mu, sigma = 0, 1
variates = nprandom.normal(mu,sigma,1000)

sns.violinplot(variates, color="pastel", vert=False);
axes.set_ylabel('x')
axis = plt.gca()
axis.xaxis.set_ticklabels(['Normal Variate'])
axes.set_title("Violin Plot of a Standard Normal Variate", weight='bold')
plt.show()
```

### Stardard Line Plot w/ Classic X and Y axis

[[Back to Top](#top)]

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(-5,6,1))
y = x**2

fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

#ax.set_xlim([-10,10])
ax.set_ylim([-25,25])

#ax.set_xticks(range(-10,10,1))
ax.set_title("$x^2$", size=24)

ax.grid(True)
ax.plot(x,y)
plt.show()
```

### Scatter Plot

[[Back to Top](#top)]

```python
from numpy import random as nprandom
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

mu, sigma = 0, 1
variate1 = nprandom.normal(mu,sigma,1000)
variate2 = nprandom.normal(mu,sigma,1000)

data_set = {"data1":variate1,"data2":variate2}
df = pd.DataFrame(data_set)

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.scatter(df["data1"].values, df["data2"].values,)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Scatter Plot", weight='bold')
plt.show()
```

### Seaborn's Joint Plot

[[Back to Top](#top)]

What's nice about Seaborn is that it leverages Pandas dataframe data structure:

```python
from numpy import random as nprandom
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # I prefer seaborn's color scheme to MATPLOTLIB's default
sns.set(font="serif") # http://stackoverflow.com/questions/23207516/matplotlib-not-listening-to-font-choices

mu, sigma = 0, 1
variate1 = nprandom.normal(mu,sigma,1000)
variate2 = nprandom.normal(mu,sigma,1000)

data_set = {"data1":variate1,"data2":variate2}
df = pd.DataFrame(data_set)

sns.jointplot("data1","data2",data=df, kind="kde");
plt.show()
```
