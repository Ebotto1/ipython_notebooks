```python
import pandas as pd
from ggplot import *
from IPython.html.widgets import interact

df = pd.read_csv('/home/pyb0k3h/Dropbox/python/ipython/notebooks/rob_story/USGS_WindTurbine_201307_cleaned.csv')
```

```python
options = ['Blade Length','Rotor Diameter','Rotor Swept Area']
@interact
def scatter_turbine(values=options):
    (ggplot(aes(x='Turbine MW', y=values), data=df) + \
        geom_point()).draw()
```

```python
import seaborn as sb

options = ['Blade Length','Rotor Diameter','Rotor Swept Area']
@interact
def sbplot(values=options):
    sb.regplot(x='Turbine MW', y=values, data=df)
```
