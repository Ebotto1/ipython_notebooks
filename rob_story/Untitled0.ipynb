```python
grouped = df.groupby('Turbine Model').size().order(ascending=False)
grouped
```

```python
models = grouped[:10].index.tolist()
models
```

```python
from ggplot import *
from IPython.html.widgets import interact

options = ['Blade Length','Rotor Diameter','Rotor Swept Area']
@interact
def scatter_turbine(values=models):
    model = df[df['Turbine Model']==values]
    (ggplot(aes(x='Turbine MW', y='Latitude-Decimal Degrees'), data=model) + \
        geom_point()).draw()
```
