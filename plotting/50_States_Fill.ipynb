# This is a Python script that charts data on top of US 48-states map

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
from matplotlib.colors import rgb2hex
from matplotlib.patches import Polygon
import operator # import this so that we can sort the data dictionary
  
# Lambert Conformal map of lower 48 states.
m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
            projection='lcc',lat_1=33,lat_2=45,lon_0=-95)

# draw state boundaries.
# data from U.S Census Bureau
# http://www.census.gov/geo/www/cob/st2000.html
shp_info = m.readshapefile(r'd:\_mycode\python\data\geographical\st99_d90','states',drawbounds=True)

# Enter the defect rate for ALL 48 states, otherwise, you will get an error
defect_rate = {
"Alaska":1.087,
"Alabama":0.059,
"Arkansas":0.456,
"Arizona":1.063,
"California":0.149,
"Colorado":0.213,
"Connecticut":0.215,
"Delaware":0.633,
"Florida":0.567,
"Georgia":0.222,
"Hawaii":0.099,
"Iowa":0.449,
"Idaho":0.000,
"Illinois":0.217,
"Indiana":0.302,
"Kansas":0.274,
"Kentucky":0.564,
"Louisiana":0.542,
"Massachusetts":0.329,
"Maryland":0.165,
"Maine":1.449,
"Michigan":0.242,
"Minnesota":0.432,
"Missouri":0.288,
"Mississippi":0.557,
"Montana":0.000,
"North Carolina":0.287,
"North Dakota":0.000,
"Nebraska":0.147,
"New Hampshire":0.000,
"New Jersey":0.198,
"New Mexico":0.272,
"Nevada":0.630,
"New York":0.073,
"Ohio":0.198,
"Oklahoma":0.559,
"Oregon":0.280,
"Pennsylvania":0.167,
"Rhode Island":0.000,
"South Carolina":0.418,
"South Dakota":0.333,
"Tennessee":0.174,
"Texas":0.707,
"Utah":0.315,
"Virginia":0.156,
"Vermont":0.000,
"Washington":0.068,
"Wisconsin":0.133,
"West Virginia":0.604,
"Wyoming":2.381
}

# choose a color for each state based on population density.
colors={}
statenames=[]
cmap = plt.cm.hot # use 'hot' colormap
vmin = 0; vmax = max(defect_rate.itervalues()) # Get the maximum value in the dictionary above

# print m.states_info[0].keys()
for shapedict in m.states_info:
    statename = shapedict['NAME']
    # skip DC and Puerto Rico.
    if statename not in ['District of Columbia','Puerto Rico']:
        pop = defect_rate[statename]
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[statename] = cmap(1.-np.sqrt((pop-vmin)/(vmax-vmin)))[:3]
    statenames.append(statename)
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance
for nshape,seg in enumerate(m.states):
    # skip DC and Puerto Rico.
    if statenames[nshape] not in ['District of Columbia','Puerto Rico']:
        color = rgb2hex(colors[statenames[nshape]]) 
        poly = Polygon(seg,facecolor=color,edgecolor=color)
        ax.add_patch(poly)
# draw meridians and parallels.
m.drawparallels(np.arange(25,65,20),labels=[1,0,0,0])
m.drawmeridians(np.arange(-120,-40,20),labels=[0,0,0,1])
plt.title('Defect Rate by By RO Dealer State\n(black=high values, white=low values)')
plt.show()

# Sort the data dictionary containing defect rates by defect rate in descending order and then print it out
defect_rate_sorted = sorted(defect_rate.iteritems(), key=operator.itemgetter(1), reverse=True)
print "CHART DATA (state:defect rate)"
rank = 1
for key, value in defect_rate_sorted:
    print "Rank", rank, key, ":" , value
    rank = rank + 1
```
