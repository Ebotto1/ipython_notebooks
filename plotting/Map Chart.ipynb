```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

# run 'help(Basemap) to see all the different kinds of maps you can draw
# This cylindrical map was drawn with bottom left-hand coordinates at (-180,10) and upper right-hand corner coordinates at (-40,75)
# This will show North America
map = Basemap(projection='cyl',llcrnrlon=-180,llcrnrlat=10, urcrnrlon=-40, urcrnrlat=75)

# draw coastlines, country boundaries, fill continents.
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color = 'coral')

# draw the edge of the map projection region (the projection limb)
map.drawmapboundary()

# draw lat/lon grid lines every 30 degrees.
map.drawmeridians(np.arange(0, 360, 10))
map.drawparallels(np.arange(-90, 90, 10))
plt.show()
```
