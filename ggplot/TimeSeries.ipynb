```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
from datetime import datetime
from datetime import date

df = pd.read_csv(r'd:\downloads\ride_share.csv',
                 converters={0: lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')})
```

```python
df.head()
```

```python
from ggplot import *

ggplot(df, aes('datetime','temp')) + \
    geom_point(color='black') + \
    geom_smooth() + \
    scale_x_date(breaks=date_breaks('1 month'), labels='%b %Y')
```
