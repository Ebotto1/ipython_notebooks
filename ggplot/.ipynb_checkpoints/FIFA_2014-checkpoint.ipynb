```python
import pandas as pd
import simplejson
from urllib import request
from pprint import pprint  # Prettier printing of json data
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)
```

**To help facilitate figuring out how to write our code, let's pretty print the json data:**

```python
pprint(matches)
```

Let's see what data type is "matches":

```python
type(matches)
```

So it is a list which means it is a list that contains nested dictionaries.  Now that we understand the structure, we can do some simple looping to print out all completed games to date.

<a id='all'></a>

## All Completed Matches To Date

[[back to top](#top)] [[today's matches](#today)] [[tomorrow's matches](#tomorrow)]

```python
for match in matches:
    if match['status'] == 'completed':
        print(match['home_team']['country'],match['home_team']['goals'])
        print(match['away_team']['country'],match['away_team']['goals'])
        print()
```

```python
json_data = request.urlopen('http://worldcup.sfg.io/matches/today')
today = simplejson.load(json_data)
```

**To help facilitate figuring out how to write our code, let's pretty print the json data:**

```python
pprint(today)
```

<a id='today'></a>

## All Today's Matches

[[back to top](#top)] [[all completed](#all)] [[tomorrow's matches](#tomorrow)]

```python
import datetime
now = datetime.datetime.now()

for match in today:
    print(match['home_team']['country'],match['home_team']['goals'])
    print(match['away_team']['country'],match['away_team']['goals'])
    print('Status:', match['status']+',', ' as of:',now.isoformat())
    print()
```

```python
json_data = request.urlopen('http://worldcup.sfg.io/matches/tomorrow')
tomorrow = simplejson.load(json_data)
```

**To help facilitate figuring out how to write our code, let's pretty print the json data:**

```python
pprint(tomorrow)
```

<a id='tomorrow'></a>

## Tomorrow's Matches

[[back to top](#top)] [[all completed](#all)] [[today's matches](#today)]

```python
for match in tomorrow:
    print(match['home_team']['country'],match['home_team']['goals'])
    print(match['away_team']['country'],match['away_team']['goals'])
    print('Start time (Brazil time):',match['datetime'])
    print()
```
