```python
import pandas as pd
from ggplot import *
import requests

ohio = pd.read_csv('/home/pyb0k3h/Dropbox/data_sets/2012-2013_School_Year_Staff_Information.csv',
                   converters={'PAY AMOUNT': lambda x: float(x.replace('$', '').replace(',',''))})
```

```python
pd.set_option('large_repr','truncate')
pd.set_option('max_columns',15)
pd.set_option('max_rows',50)
ohio
```

```python
criterion1 = ohio['PAY AMOUNT'] > 0
criterion2 = ohio['DAYS WORKED'] > 180
criterion3 = ohio['HOURS WORKED PER DAY'] >=8
ohio_fulltime = ohio[criterion1 & criterion2 & criterion3]
ohio_fulltime
```

```python
criterion1 = ohio['PAY AMOUNT'] > 0
criterion2 = ohio['DAYS WORKED'] > 180
criterion3 = ohio['HOURS WORKED PER DAY'] >=8
criterion4 = ohio['JOB DESCRIPTION'] == 'Teacher Assignment'
ohio_teachers = ohio[criterion1 & criterion2 & criterion3 & criterion4]
ohio_teachers
```

```python
ggplot(aes(x='PAY AMOUNT'), data=ohio_teachers) + \
    ggtitle("Ohio Full-Time Teachers") + \
    ylab("Qty") + \
    geom_histogram()
```

```python
criterion1 = ohio['PAY AMOUNT'] > 20000
criterion2 = ohio['DAYS WORKED'] > 180
criterion3 = ohio['HOURS WORKED PER DAY'] >=8
criterion4 = ohio['JOB DESCRIPTION'] == 'Teacher Assignment'
criterion5 = ohio['COUNTY NAME'] == 'Franklin'
ohio_franklin = ohio[criterion1 & criterion2 & criterion3 & criterion4 & criterion5]
ohio_franklin
```

```python
ggplot(aes(x='PAY AMOUNT'), data=ohio_franklin) + \
    ggtitle("Ohio Full-Time Teachers") + \
    ylab("Qty") + \
    geom_histogram()
```

```python
ggplot(aes(x='PAY AMOUNT'), data=ohio_franklin) + \
    ggtitle("Ohio Full-Time Teachers") + \
    facet_wrap('DISTRICT NAME') + \
    geom_histogram()
```
