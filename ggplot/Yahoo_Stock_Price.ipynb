```python
from ggplot import *
import pandas.io.data as web
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

ticker = 'HMC'
start = datetime(2014, 10, 1)
end = datetime(2014, 10, 31)

df=web.DataReader(ticker, 'yahoo', start, end)
```

```python
df.head()
```

But I need to make the index to be a column and also need to drop Volume column since I will plot the Open, High, Low, Close, and Adj Close prices

```python
# Converts the index into a column
df.reset_index(level=0, inplace=True)
```

```python
df.head()
```

```python
# Drops the Volume column
df.drop('Volume', axis=1, inplace=True)
```

```python
df.head()
```

But if my objective is to plot the Open, High, Low, Close, and Adj Close on the same plot, I need to transform this dataframe to the long form using the melt() method.  Basically, it converts the columns to row data.

```python
df_long = pd.melt(df, id_vars=['Date'])
```

```python
df_long
```

## Now we're ready for some ggplot!

```python
stock_plot = ggplot(df_long, aes(x='Date', y='value', colour='variable')) + \
    geom_line() + \
    ggtitle('Stock Ticker: ' + ticker) + \
    ylab('Price (US$)')
    
fig = stock_plot.draw()
ax = fig.gca()
labels = ax.get_xticklabels() 
for label in labels: 
    label.set_rotation(270)
plt.show()
```
