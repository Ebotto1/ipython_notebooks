## Charting a single stock ticker

```python
from ggplot import *
import pandas.io.data as web
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

ticker = 'HMC'
start = datetime(2014, 10, 1)
end = datetime(2014, 11, 7)

df=web.DataReader(ticker, 'yahoo', start, end)
```

```python
df.head()
```

But I need to make the index to be a column and also need to drop Volume column since I will try to plot the Open, High, Low, Close, and Adj Close prices

```python
# Converts the index into a column and the new index will be 0...n
df.reset_index(level=0, inplace=True)
```

```python
df.head()
```

```python
# Drops the Volume column
df.drop('Volume', axis=1, inplace=True)
```

```python
df.head()
```

But if my objective is to plot the Open, High, Low, Close, and Adj Close on the same plot, I need to transform this dataframe to the long form using the melt() method.  Basically, it converts the columns to row data.

```python
df_long = pd.melt(df, id_vars=['Date'])
```

```python
df_long
```

## Now we're ready for some ggplot!

```python
stock_plot = ggplot(df_long, aes(x='Date', y='value', colour='variable')) + \
    geom_line() + \
    ggtitle('Stock Ticker: ' + ticker) + \
    ylab('Price (US$)')
    
fig = stock_plot.draw()
ax = fig.gca()
labels = ax.get_xticklabels() 
for label in labels: 
    label.set_rotation(270)
plt.show()
```

## Charting multiple stock tickers

[Back to Top](#top)

```python
from ggplot import *
import pandas.io.data as web
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stocks = {'tech':['GOOG', 'MSFT', 'LNKD', 'YHOO', 'FB','HPQ','AMZN'],
          'auto':['TM','F','HMC','NSANY',]
          }
```

```python
start_date = datetime(2012,1,1)
end_date = datetime(2014, 11, 7)

def get_px(stock, start, end):
    return web.get_data_yahoo(stock, start, end)['Adj Close']

df = pd.DataFrame({n: get_px(n, start_date, end_date) for n in stocks['auto']})
```

```python
df.head()
```

```python
df.reset_index(level=0, inplace=True)
```

```python
df.head()
```

```python
df_long = pd.melt(df, id_vars=['Date']).dropna()
```

```python
df_long = df_long.rename(columns = {'Date':'Date','variable':'Stock','value':'Price'})
```

```python
df_long.head()
```

### One advantage of using ggplot is, you can change your date intervals very easily.  With MATPLOTLIB, it is more verbose.

```python
stock_plot = ggplot(df_long, aes(x='Date', y='Price', colour='Stock')) + \
    geom_line() + \
    ylab('Price (US$)') + \
    ggtitle('Adj Close') + \
    scale_x_date(labels='%b %d %y',breaks=date_breaks('year') )
    
# Have to use MATPLOTLIB api to rotate x tick labels.  Hopefully some day, ggplot will implement a rotation option.
fig = stock_plot.draw()
ax = fig.gca()
labels = ax.get_xticklabels() 
for label in labels: 
    label.set_rotation(270)
plt.show()
```
