```python
from lxml import html
import urllib.request as request

html_page = html.fromstring(request.urlopen('http://www.carcomplaints.com/Honda/').read())
```

```python
print(html.tostring(html_page, pretty_print=True).decode('utf-8'))  # Due to us using Python 3, we have to decode from bytes to string
```

#### To get the title, we select the element we want using cssselect() method, which returns a list of html elements

```python
title = html_page.cssselect('title')[0]
```

```python
title.text_content()
```

```python
meta = html_page.cssselect('meta')[0]
```

```python
meta.keys()
```

```python
meta.values()
```

```python
columns = html_page.find_class("column bar")
```

```python
for column in columns:
    for li in column.cssselect("a"):
        print(li.text_content())
```

```python
for column in columns:
    for li in column.cssselect("span"):
        print(li.text_content())
```
