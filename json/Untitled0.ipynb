```python
from prettytable import PrettyTable
import simplejson
from urllib import request
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

complete_events = PrettyTable(["Match Number", "Team", "Player", "Time", "Type Of Event"])

for match in matches:
    if match['status'] == 'completed':
        home_team = match['home_team']['country']
        away_team = match['away_team']['country']
        match_number = match['match_number']
        for home_event in match['home_team_events']:
            player = home_event['player']
            time = home_event['time']
            event = home_event['type_of_event']
            complete_events.add_row([match_number, home_team, player, time, event])
            
        for away_event in match['away_team_events']:
            player = away_event['player']
            time = away_event['time']
            event = away_event['type_of_event']
            complete_events.add_row([match_number, away_team, player, time, event])

print(complete_events)
```

```python
column_names = complete_events._get_field_names()
column_names
```

```python
row_data = complete_events._rows
row_data
```

```python
import csv
from io import StringIO

output = StringIO()

writer = csv.writer(output)
writer.writerows([column_names])
writer.writerows(row_data)
events = output.getvalue()
output.close()
```

### Limit the dataframe to just goals

```python
df_goals = df[(df['Type Of Event']=='goal') | (df['Type Of Event']=='goal-penalty')]
```

### Top Goal Points by Player

Now that I have limited my data to include only goals, I can do goal counts.  Let's see which players have the most goals:

```python
player = df_goals.groupby(['Player'])['Type Of Event'].count()
player.sort(ascending=False)
player
```

**But I realized, it would be nice to know which country they are from, so I'll group the counts by team and then by player:**

### Top Goal Counts by Player (with team name)

```python
grouped = df_goals.groupby(['Team', 'Player'])['Type Of Event'].count()
grouped.sort(ascending=False)
grouped
```

### Top Goal Counts by Team/Country

```python
from prettytable import PrettyTable
import simplejson
from urllib import request
json_data = request.urlopen('http://worldcup.sfg.io/matches')
matches = simplejson.load(json_data)

completed_matches = PrettyTable(["Match Number", "Home Team","Home Team Goals", "Away Team", "Away Team Goals", "Winner"])

for match in matches:
    if match['status'] == 'completed':
        match_number = match['match_number']
        home_team = match['home_team']['country']
        away_team = match['away_team']['country']
        home_team_goals = match['home_team']['goals']
        away_team_goals = match['away_team']['goals']
        winner = match['winner']
        completed_matches.add_row([match_number, home_team, home_team_goals, away_team, away_team_goals, winner])

print(completed_matches)
```

```python
column_names = completed_matches._get_field_names()
column_names
```

```python
row_data = completed_matches._rows
row_data
```

```python
import csv
from io import StringIO

output = StringIO()

writer = csv.writer(output)
writer.writerows([column_names])
writer.writerows(row_data)
contents = output.getvalue()
output.close()
```

```python
away_team = df[['Away Team', 'Away Team Goals']]
away_team.rename(columns={'Away Team':'Home Team', 'Away Team Goals':'Home Team Goals'}, inplace=True)
away_team
```

```python
pieces = [home_team, away_team]
home_and_away = pd.concat(pieces)
home_and_away.index = range(0,len(home_and_away))
home_and_away.rename(columns={'Home Team':'Team', 'Home Team Goals':'Goals'}, inplace=True)
home_and_away
```
