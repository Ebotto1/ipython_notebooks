### If you rather observe the columns vertically, we can transpose the dataframe:

```python
df1 = vehicles.head(1)
df1.transpose()
```

```python
vehicles.make.value_counts()
```

```python
criteria1 = vehicles.make.isin(['Toyota','Honda','Nissan','Hyundai','Kia','Mazda','Acura','Lexus'])
major_asian_makers = vehicles[criteria1]
```

```python
help(vehicles.pivot_table)
```

```python
df2 = major_asian_makers.pivot_table(values='comb08',index=['year'],columns=['make'],fill_value=0,aggfunc='count')
```
