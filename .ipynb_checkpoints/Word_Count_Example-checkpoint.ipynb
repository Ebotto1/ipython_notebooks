```python
import string
from collections import Counter
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

bunch_of_words = """
customer states engine is too loud
customer stated the engine is noisy
customer states engine is making a sound
customer can hear a loud bang
customer complains of a noise
customer complaints of noise
not work, not working, no work,
broke, break, brake,
judder,judders, juddering,
rattle, rattles, rattled,
engine rattle, eng noise, engine vibrates, engine vibration, engine vibrating
"""

bunch_of_words = bunch_of_words.lower()

for character in bunch_of_words:
    if character in string.punctuation:  # string.punctuation = !"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~
        bunch_of_words = bunch_of_words.replace(character," ")
        
tokenizer = RegexpTokenizer("[\w']+")
words_list = tokenizer.tokenize(bunch_of_words)
english_stops = set(stopwords.words('english'))
custom_stop_words = ['custom','state','client','complain','complaint','said']

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

final_words_list = [word for word in words_list if word not in english_stops]

final_words_list_stemmed = [stemmer.stem(word) for word in final_words_list]

final_words_list_lemmatized = [lemmatizer.lemmatize(word) for word in final_words_list]

final_words_list.sort()
final_words_list
```
